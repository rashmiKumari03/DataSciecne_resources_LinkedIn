##### Retail 

```python
# Get quantities from the customer and multiple it with its price.

quantity1 = int(input("Enter the quantity of the first item: "))
quantity2 = int(input("Enter the quantity of the second item: "))

# Calculate the subtotal  #Suppose the price of 1 unit of first item is $20 and 1 unit of second item is $50.
subtotal = (quantity1 * 20) + (quantity2 * 50)

# Determine the discount based on the subtotal
if subtotal > 200:
    discount = 0.2 * subtotal    #0.2 means 20% discount
elif subtotal > 100:
    discount = 0.1 * subtotal    #0.1 means 10% discount
else:
    discount = 0

# Calculate the tax and total
tax = 0.1 * subtotal
total = subtotal - discount + tax

# Display the results
print(f"Subtotal ${subtotal:.2f}")
print(f"Discount ${discount:.2f}")
print(f"Tax ${tax:.2f}")
print(f"Total ${total:.2f}")
```

#### Healthcare

```python
# Get the patient's age, sex, and cholesterol level
age = int(input("Enter the patient's age: "))
sex = input("Enter the patient's sex (M or F): ")
cholesterol = int(input("Enter the patient's cholesterol level: "))

# Cholestrol level: Normal: Less than 200 mg/dL. Borderline high: 200 to 239 mg/dL. High: At or above 240 mg/dL.

# Calculate the risk of heart disease based on the patient's characteristics
if sex == "M" and age > 45 and cholesterol > 200:
    risk = "high"
elif sex == "F" and age > 55 and cholesterol > 220:
    risk = "high"
elif sex == "M" and age > 35 and cholesterol > 150:
    risk = "moderate"
elif sex == "F" and age > 45 and cholesterol > 180:
    risk = "moderate"
else:
    risk = "low"

# Display the result
print(f"The patient's risk of heart disease is {risk}.")
```

```python
# Check the creditworthiness of a borrower and approve or deny a loan

credit_score = int(input("Enter the credit score:"))
income = int(input("Enter the annual income: "))
employment_status = input("Enter the employment status (employed or unemployed):")

if credit_score >= 700 and income >= 50000:
    if employment_status == "employed":
        loan_approval = "approved"
    else:
        loan_approval = "denied"
else:
    loan_approval = "denied"

print(f"The loan approval status is {loan_approval}.")
```

##### Insurance

```python
# Calculate the premium based on the risk profile of the insured

age = int(input("Enter the age: "))
gender = input("Enter the gender (M or F): ")
smoking_status = input("Enter the smoking status (yes or no): ")

#In the insurance industry, smoking is considered a risk factor for various health conditions, such as cancer heart disease,
#and respiratory problems.Therefore, insurance companies often charge higher premiums to smokers to account for the increased 
#risk.

if age >= 25 and age <= 60:
    if gender == "M":
        premium = 1000
    else:
        premium = 900

    if smoking_status == "yes":
        premium = premium * 1.2
else:
    premium = "not eligible"

print(f"The premium is {premium}.")
```

##### Banking 

```python
# Check the account balance and apply fees if necessary
account_balance = int(input("Enter the account balance: "))

#What is Overdraft_fee
#Suppose you have $50 in your account, but you need $75 to pay for a sudden expense, such as a car repair. 
#You are $25 short, but the bank loans you that $25 and the payment clears. Let's say the bank charges a $30 
#overdraft fee. That $30 fee plus the $25 overdraft would leave you with an account balance of -$55.

if account_balance < 0:
    overdraft_fee = 35
    new_balance = account_balance - overdraft_fee
else:
    new_balance = account_balance

print(f"The new account balance is: {new_balance}.")
```
