`"Functions": Reuseable Block of Code.`

A function in Python is like a recipe that we can reuse over and over again. 
It is a block of code that performs a specific task. Functions help us organize our code and make it easier to understand.

The benefits of using functions are:
- They make our code modular and break it into smaller, manageable pieces.
- We can use the same function multiple times, saving us from writing the same code again and again.
- Functions have meaningful names, which makes our code easier to read and understand.
- They make our code easier to maintain and update.
- Functions provide a way to focus on the main task without worrying about the details.

To define a function, we use the 'def' keyword followed by the function name and parentheses. Inside the function, we write the code that we want to execute. We can also pass values to the function using parameters, and the function can return a value if needed.

When we want to use a function, we simply call it by writing its name followed by parentheses. We can provide the required values inside the parentheses if the function expects any.

In summary, `functions` in Python are `reusable blocks of code` that help us organize and simplify our program. They make coding easier, promote code reuse, and improve the overall readability and maintainability of our code.

Modular Coding : It involves separating a program's functions into independent pieces or building blocks, each containing all the parts needed to execute a single aspect of the functionality.

Be careful we can use extend to unpack the element but the extend() method, which expects an iterable as an argument, but you're passing a single value (i) instead. To fix the error, we can change extend(i) to append(i) since we want to add the individual values to the lists.

````python
#Tried just for integer,float,string and boolean 

def datatype_separation(*args):
    integer=[]
    floats=[]
    strings=[]
    boolean=[]
    comp=[]
    
    
    for i in args:
        if type(i)==int:
            integer.append(i)
        elif type(i)==float:
            floats.append(i)
        elif type(i)==str:
            strings.append(i)
        elif type(i)==bool:
            boolean.append(i)
        elif type(i)==complex:
            comp.append(i)
        
        
    return integer,floats,strings,boolean,comp
````

````python
def datatype_separation(*args):
    integer=[]
    floats=[]
    strings=[]
    boolean=[]
    lists=[]
    sets=[]
    tuples=[]
    dicts=[]
    comp=[]
    
    for i in args:
        if type(i)==int:
            integer.append(i)
        elif type(i)==float:
            floats.append(i)
        elif type(i)==str:
            strings.append(i)
        elif type(i)==bool:
            boolean.append(i)
        elif type(i)==complex:
            comp.append(i)
        elif type(i)==list:
            lists.append(i)
        elif type(i)==tuple:
            tuples.append(i)
        elif type(i)==set:
            sets.append(i)
        elif type(i)==dict:
            dicts.append(i)
       
        
    return integer,floats,strings,boolean,comp,lists,tuples,sets,dicts
````

**TRY IT:👩‍💻Dymanic Way to create a function that if datatype not found automatically new empty list must be created.**

````python
def datatype_separation(*args):
    data_types = {}    #Creating an empty dictionary.

    for i in args:
        found_type = False

        for data_type, data_list in data_types.items():
            if type(i)== data_type:
                data_list.append(i)
                found_type = True
                break
        
        if not found_type:
            data_types[type(i)] = [i]

    # return data_types  we can also return datatype and values using return data_types .
    
    return tuple(data_types.values())
````

````python
result = datatype_separation(10, 3.14, 'hello', True, [1, 2, 3], 'world', (4, 5), {1, 2, 3}, {'a': 1, 'b': 2},2+9j)
print(result)
````

**Explanation:**



```python
def datatype_separation(*args):
    data_types = {}
```

- This is the function definition. It takes in variable-length arguments `args` and initializes an empty dictionary `data_types` to store the lists for each data type encountered.

```python
    for i in args:
        found_type = False

        for data_type, data_list in data_types.items():
            if type(i) == data_type:
                data_list.append(i)
                found_type = True
                break

        if not found_type:
            data_types[type(i)] = [i]
```

- This is the main logic of the function. It iterates over each argument `i` in the `args` tuple.
- The variable `found_type` is initialized as `False` to keep track of whether the data type of `i` is already present in the `data_types` dictionary.
- It then iterates over each key-value pair in the `data_types` dictionary using the `items()` method.
- Inside this loop, it checks if the type of `i` (`type(i)`) matches the current data type (`data_type`) from the `data_types` dictionary.
  - If there's a match, it means the data type is already present, so it appends `i` to the corresponding list (`data_list`), sets `found_type` to `True`, and breaks out of the inner loop.
- If no match is found in the inner loop (i.e., `found_type` is still `False`), it means the data type is not yet present in the `data_types` dictionary. In this case, it creates a new entry in the dictionary with the data type as the key and a list containing `i` as the value.

```python
    return tuple(data_types.values())
```

- Finally, the function returns a tuple containing all the lists from the `data_types` dictionary.

The function essentially separates the input arguments into different lists based on their data types, and the resulting lists are returned as a tuple.

````python
def conc_list_only(*args,**kwargs):
    c=[]
    for i in args :
        if type(i)==list:
            c.extend(i)
            
    for i in kwargs.values():    #Because we want values not key here from kwargs.
        if type(i)==list:
            c.extend(i)
            

    return c
````
