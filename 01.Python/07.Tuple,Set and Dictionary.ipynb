Tuple in Python:
- A tuple is an ordered sequence of elements, enclosed in parentheses `()`.
- Tuples are immutable, meaning their elements cannot be modified once the tuple is created.
- Elements in a tuple can be of different data types (e.g., integers, strings, floats).
- Tuples support indexing and slicing to access individual elements or subsequences.
- Tuples can be created using the tuple() constructor or by simply enclosing elements within parentheses.

Applications of Tuples:
- Multiple return values: Tuples are commonly used to return multiple values from a function.
- Unpacking: Tuples are useful for unpacking elements into multiple variables.
- Immutable data: Tuples can be used to store constant or unchangeable data.
- Dictionary keys: Tuples can be used as keys in dictionaries since they are immutable.
- Data integrity: Tuples can be used to ensure data integrity by creating records that cannot be modified.

Similarities between Lists and Tuples:
- Both lists and tuples are used to store and organize multiple elements.
- Both support indexing and slicing to access elements or subsequences.
- Elements in both lists and tuples can be of different data types.
- Both can be nested, allowing for the creation of multidimensional data structures.
- Both can be iterated over using loops or comprehensions.

Differences between Lists and Tuples:
- Lists are mutable, meaning their elements can be modified, added, or removed. Tuples are immutable and cannot be modified.
- Lists are represented by square brackets `[]`, while tuples are represented by parentheses `()`.
- Lists have more built-in methods for manipulation, such as append(), extend(), and remove(). Tuples have fewer methods due to their immutability.
- Lists are generally used when you need a collection that can be modified, while tuples are used when you need an immutable collection or want to ensure data integrity.

In summary, tuples are immutable collections, commonly used for multiple return values, unpacking, and storing constant data. Lists, on the other hand, are mutable collections that allow for dynamic modifications and are often used when you need flexibility in adding, removing, or modifying elements.

* NOTE:
  In Python, a tuple is an immutable data type, which means its contents cannot be changed after it is created. However, if a mutable object, such as a list, is contained within a tuple, it is possible to modify the elements of that mutable object. It's important to note that modifying the mutable object inside the tuple does not change the immutability of the tuple itself.

  When accessing and modifying an element of a list inside a tuple, you are modifying the list object, not the tuple directly. This means that the identity (or id()) of the tuple remains the same before and after the modification, while the mutable object (the list) inside the tuple gets modified.

**`Highlighting the mutability of lists and the immutability of tuples:`**

`List in Tuple:`
- In Python, a list can be an element within a tuple.
- Tuples are immutable, meaning they cannot be changed once created.
- However, if a list is an element within a tuple, the list itself remains mutable.
- This means that the elements of the list can be modified, added, or removed.
- Modifying the list does not affect the immutability of the tuple containing it.
- The identity of the tuple (its memory address) remains the same before and after modifying the list.

`Tuple in List:`
- In Python, a tuple can be an element within a list.
- Lists are mutable, allowing for changes to their elements.
- However, tuples are immutable and cannot be modified once created.
- Even if a tuple is an element within a list, the tuple itself remains immutable.
- The elements within the tuple cannot be modified, added, or removed.
- Modifying the list itself, such as adding or removing elements, is possible.

In summary, the mutability of lists allows for modifications to their elements, even when they are elements within tuples. On the other hand, the immutability of tuples prevents direct modifications to their elements, even if the tuples themselves are elements within lists.

**👩‍💻`REAL WORLD APPLICATION OF TUPLE:`**

1. `Database Operations`: Tuples are commonly used to represent rows of data in databases. Each tuple corresponds to a record, with elements representing specific attributes. Tuples offer efficient storage and retrieval of data, making them essential for managing databases effectively.


2. `Geographic Coordinates`: Tuples, typically in the form of (latitude, longitude), are used to pinpoint locations on the Earth's surface. This representation is crucial for mapping systems, GPS navigation, and geolocation services, allowing accurate positioning and navigation.


3. `Financial Transactions`: Tuples are employed to store information related to financial transactions. A tuple may contain details such as transaction ID, date and time, amount, and parties involved. Tuples enable efficient organization, tracking, analysis, and auditing of financial data in various financial systems.


4. `Sensor Readings`: In domains like IoT and environmental monitoring, sensors collect data from the physical world. Tuples are commonly used to store sensor readings, with each element representing different parameters like temperature, humidity, pressure, or light intensity. Tuples facilitate the storage, processing, and analysis of sensor data in these applications.


5. `Game Development`: Tuples play a significant role in game development. They can be used to represent game states, store player information such as position, score, health, and manage game events. Tuples provide a structured and efficient way to handle and manipulate game-related data.


6. `PIN Verification`: Tuples can store a combination of a username and a hashed password in a database. During the login process, the entered PIN gets hashed and compared against the stored hashed password to authenticate the user. Tuples are employed to associate the username with the corresponding hashed password securely.


7. `Password Strength Checking`: Tuples are useful for evaluating the strength of a password. They can store attributes of a password, such as length, presence of uppercase and lowercase characters, digits, and special characters. By examining these attributes, one can determine the strength of the password and provide appropriate feedback to the user, ensuring stronger and more secure passwords.

**📀INTRODUCTION**

In Python, a dictionary is a built-in data type that represents an unordered collection of key-value pairs. It is also known as an associative array, map, or hash table in other programming languages. Dictionaries are widely used because they provide an efficient way to retrieve and store data based on a unique key.

Key features of dictionaries in Python include:

- **Key-Value Pairs**: Dictionaries consist of key-value pairs, where each key is unique and associated with a corresponding value. The key serves as an identifier or label for the value.

- **Unordered**: The elements in a dictionary are not stored in a specific order. This means that the order in which the key-value pairs are inserted may not be maintained when you iterate over the dictionary or print its contents.

- **Mutable**: Dictionaries are mutable, allowing you to add, remove, or modify key-value pairs after creation.

- **Dynamic Size**: Dictionaries can dynamically change their size as key-value pairs are added or removed. They do not have a fixed length.

- **Key-Based Access**: Values in a dictionary are accessed by their corresponding keys. This allows for efficient retrieval of values, as dictionaries use a hash function to map keys to their associated values.

- **Flexible Key Types**: Keys in a dictionary can be of various data types, such as strings, integers, or even tuples (as long as they are immutable).

Dictionaries are useful in situations where you need to store and retrieve data based on a unique identifier or label. They are commonly used for tasks like:

- Storing and accessing configuration settings.
- Creating lookup tables or mapping between values.
- Counting occurrences of elements in a dataset.
- Grouping related data together.

Overall, dictionaries provide a powerful and flexible way to organize and manipulate data in Python by associating values with unique keys.

```python
#Q1: TAKE INPUT AS A PARAGRAPH BY USING INPUT FUNCTION.
#Ans1:

paragraph=input("Enter Something About Buiness.")
```

```python
#Q2: CONVERT ALL OF THIS INPUT INTO LIST OF WORDS.
#Ans2:

para_string=paragraph.split(' ')
```

```python
para_string
```

```python
# Datatype of this para_string.
type(para_string)
```

```python
#Q3. REMOVE ALL THE DUPLICATE ELEMENTS FROM YOUR LIST.
#Ans3:

l=set(para_string)    #These all words just comes once.All of them are unique as i have used set()
l
```

```python
#Q4. USE ABOVE DISTINCT ELEMENT AS A KEY FOR DICTIONARY AND ASSIGN YOUR NAME AS A VALUE OF ALL THE KEY FOR DICTIONARY.
#Ans4:

for i in l:
    print(i,end=' ')
```

```python
value="rashmi"
```

```python
dictionary={}              #Empty dictionary

for i in l:

    dictionary[i]=value   #This is mapping the value to each of the word in l. i containes the word so...i--->value.
                         #And getting mapped in each iternation.
    
print(dictionary)
```

```python
#Q5.TRY TO PRINT TYPE OF ALL THE VALUES.
#Ans5:

for i in dictionary.values():
    print(i,'--->',type(i))
```

```python
#Q6.TRY TO PRINT TYPE OF ALL THE Keys.
#Ans5:

for i in dictionary.keys():
    print(i,'--->',type(i))
```

```python
d={"Company":"Sky is the limit",
  "Course":"Full Stack Data Science",
  "Year":2023}
```

```python
for i in d:
    print("Keys is:",i,'  ----->  ',"Value is:",d[i])
```

```python
"Course" in d
```

```python
for i in d:
    if i=='Year':
        if d[i] > 2020:
            print("Year is Fine")
        else:
            d[i]=2024
```

```python
#If course name length is more than 20.split each of the word and store in list.

for i in d.keys():
    
    if i=='Course':
        print("Value associated with Course:",d[i])
        print("Length of",d[i],'is:',len(d[i]))
        
        if len(d[i])>20:
            l=d[i].split(' ')
            print(l)
```

```python
#We can say Sets are Mutable because we can add or remove elements from it.

# Create a set
my_set = {1, 2, 3, 4}

# Add an element to the set
my_set.add(5)
print(my_set)  # Output: {1, 2, 3, 4, 5}

# Remove an element from the set
my_set.remove(3)
print(my_set)  # Output: {1, 2, 4, 5}
```

| Feature         | Lists                   | Tuples                    | Sets                      | Dictionary                |
|-----------------|-------------------------|---------------------------|---------------------------|---------------------------|
| Mutable         | Yes                     | No                        | Yes                       | Yes                       |
| Order           | Ordered                 | Ordered                   | Unordered                 | Unordered                 |
| Duplicates      | Allowed                 | Allowed                   | Not allowed               | Not allowed               |
| Indexing        | Yes                     | Yes                       | No                        | Yes (via keys)            |
| Modification    | Add, remove, modify     | Immutable (unchangeable)  | Add, remove               | Add, remove, modify       |
| Length          | Dynamic                 | Fixed                     | Dynamic                   | Dynamic                   |
| Use cases       | Mutable sequences       | Immutable sequences       | Unique elements, operations | Key-value mappings        |
| Iteration       | Efficient               | Efficient                 | Efficient                 | Efficient                 |
| Membership test | Efficient               | Efficient                 | Efficient                 | Efficient                 |
