**extend():


|                      | `append()`                                              | `extend()`                                                |
|----------------------|---------------------------------------------------------|-----------------------------------------------------------|
| Purpose              | Add a single element to the end of a list                | Add multiple elements from an iterable to the end of a list |
| Number of arguments  | 1                                                       | 1                                                         |
| Argument type        | Single element                                          | Iterable                                                  |
| Effect               | Appends the element as a single item to the list         | Appends individual elements from the iterable to the list |
| List modification    | In-place modification of the list                       | In-place modification of the list                         |
| Example              | `my_list.append(4)`                                     | `my_list.extend(another_list)`                            |
| Result               | `[1, 2, 3, 4]`                                         | `[1, 2, 3, 4, 5, 6]`                                      |


````python
l1=[1,2,3,4,5]
l2=[2,3,4,5,6]
````

````python
l1.extend(l2)
````

````python
print(l1)  #Therfore, l1 got extended and all the elements got appended into the l1 list.
````

````python
print(l2)
````

### Difference Between insert(),append() and extend()


|                      | `insert()`                                                     | `append()`                                                | `extend()`                                                           |
|----------------------|----------------------------------------------------------------|-----------------------------------------------------------|----------------------------------------------------------------------|
| Purpose              | Insert a single element at a specific index position             | Append a single element to the end of a list               | Append multiple elements from an iterable to the end of a list       |
| Number of arguments  | 2                                                              | 1                                                         | 1                                                                    |
| Argument types       | Index position, Single element                                   | Single element                                           | Iterable object                                                      |
| Effect               | Inserts the element at the specified index                       | Appends the element as a single item to the list          | Appends individual elements from the iterable to the list             |
| List modification    | In-place modification of the list                                | In-place modification of the list                       | In-place modification of the list                                   |
| Iterable object      | No                                                             | No                                                        | Yes                                                                  |
| Unwrapping of elements | Not applicable                                                | Not applicable                                           | Elements are unwrapped individually from the iterable                |
| Example              | ```python my_list = [1, 2, 3] my_list.insert(1, 4) print(my_list) ``` | ```python my_list = [1, 2, 3] my_list.append(4) print(my_list) ``` | ```python my_list = [1, 2, 3] another_list = [4, 5, 6] my_list.extend(another_list) print(my_list) ``` |

````python
#Q4. If your Nested list contain string then , Find out its index in nested list and print it,then remove the string.
````

````python
l = [[1, 2], [1, 3], [43, 21.2, 'xys'], 'GDP', 'INR', 9.9,['YES',True]]

for i in l:
    if type(i)==list:
        j=0
        while j < len(i):
            if type(i[j]) is str:
                print("Index of the string {} is:".format(i[j]),j)
            j=j+1
            
        j=0
        while j<len(i):
            if type(i[j]) is str:
                print("The string in the nested list is {}".format(i[j]))
                i.remove(i[j])
                continue
                
            j=j+1
            
print("List after removal of string from nested list is:\r",l,'\n')


#Note: We want to remove only those string , which are there in nested list only.
````

**Get input() from the user in a list.**

````python
l=[]    #Creating an empty list.

for i in range(5):
    print("Enter the elements you want to append in a list at index",i,':')
    l.append(input())
print(" required List is:\n ",l)
````

````python
#We can see everything here is in form of string inside the List.Therefore we can improve it.
````

````python
l = []  # Creating an empty list.

for i in range(5):
    print("Enter the element you want to append in the list at index", i, ':')
    data_type = input("Enter the data type (int, float, bool, list): ")
    value = input("Enter the value: ")

    if data_type == 'int':
        l.append(int(value))
    elif data_type == 'float':
        l.append(float(value))
    elif data_type == 'bool':
        l.append(bool(value))
    elif data_type == 'list':
        l.append(eval(value))
    else:
        print("Invalid data type. Skipping element.")

print("Required list is:\n", l)
````

* `enumerate()` is a built-in Python function that allows you to iterate over a sequence (such as a list, tuple, or string) while also keeping track of the index or position of each element.
